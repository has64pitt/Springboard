#### Create a log of the price to make the analysi easier

```python
listings_df = pd.read_csv('./data/listings_clean.csv')
listings_df['host_since'] = pd.to_datetime(listings_df['host_since'])

cols = ['price','security_deposit','cleaning_fee']
for col in cols:
    listings_df[col] = listings_df[col].apply(lambda x: float(x[1:].replace(',','')))
    listings_df[col] = listings_df[col].astype(float)

listings_df = listings_df.loc[listings_df.price>0]
listings_df['price_log'] = listings_df['price'].apply(lambda x: log10(x))    
listings_df['cleaning_fee_log'] = listings_df['cleaning_fee'].apply(lambda x: log10(x+1))
listings_df['security_deposit_log'] = listings_df['security_deposit'].apply(lambda x: log10(x+1))
listings_df["Point"] = listings_df.apply(lambda row:  Point(row["longitude"],     row["latitude"]), axis=1)


print(listings_df.shape)
listings_df.head(5)
```

### Next, I will plot each listing based on its location and overlay with map, so we see it on a map. The listings are grouped by zipcode, and a mean price will be cacluated to color code the area.

#### Prepare the dataframe with zipcode and boudnary map

```python
austin_map = gpd.read_file("./data/neighbourhoods.geojson")
austin_map["zipcode"] = austin_map["neighbourhood"].astype(int)
austin_map.drop(['neighbourhood_group','neighbourhood'],axis=1,inplace=True)
```

#### Create listings dataframe with geodataframe format

```python
listings_df2 = gpd.GeoDataFrame(listings_df, geometry="Point")
```

#### Merge zipcode dataframe (with boundary infomation) with the price dataframe, price dataframe is the mean for each zipcode

```python
price_df = listings_df.pivot_table(index='zipcode',values='price',aggfunc='mean')
price_df['log_mean_price'] = price_df['price'].apply(lambda x: log10(x))
price_df = pd.merge(austin_map, price_df, how="left", on="zipcode")
price_df = price_df.set_index("zipcode")
price_df.head()
```

```python
fig, ax = plt.subplots(figsize=(20,20))

base = price_df.plot(column="price", cmap="Reds", \
                     edgecolor="whitesmoke", ax=ax, legend=False)
base.axis("off")

price_df.apply(lambda x: ax.annotate(s=x.name, \
                                     xy=x.geometry.centroid.coords[0], \
                                     ha="center", fontweight="light",  \
                                     fontsize=12, color="black"), axis=1)


min_price = price_df["price"].min()
max_price = price_df["price"].max()

gradient = plt.cm.ScalarMappable(cmap="Reds", \
                                 norm=plt.Normalize(vmin=min_price, \
                                                    vmax=max_price))
print(gradient)
gradient._A = []
cbar = fig.colorbar(gradient, fraction=0.04, pad=.01)

listings_df2.plot(ax=base, cmap="winter", marker="o", markersize=3, \
                 alpha=.8, column="host_is_superhost", \
                 categorical=True, legend=True)


cbar.ax.set_yticklabels(["$" + i.get_text() for i in cbar.ax.get_yticklabels()])

legend = ax.get_legend()
legend.set_bbox_to_anchor((0.25, 0.75, 0.10, 0.10))

plt.title("Average Airbnb Price per Zipcode in Austin, TX", \
          fontdict={"fontsize":20})

plt.show()
```

### It is obvious that the most expensive listings are in the west of Austin, and downtown Austin area has more listings than other areas
